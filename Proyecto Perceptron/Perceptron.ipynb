```python
import matplotlib.pyplot as plt
import random

w1=random.random()
w2=random.random() 
b=random.random()
nombre=""

    
def funcionActivacion(x1,x2):
    n = ((w1*x1) + (w2*x2)) + b
    if (n>=0):
        return 1
    else:
        return 0
    
def actualizacionPesos(a,t,x1,x2):
    global w1
    global w2
    global b
    w1 = w1 + 0.3*(t-a)*x1
    w2 = w2 + 0.3*(t-a)*x2
    b = b + 0.3*(t-a)

def AND():  
    flag=0
    while flag==0:
        flag = 1
        if funcionActivacion(0,0)!=0:
            flag=0
            actualizacionPesos(1,0,0,0)
        if funcionActivacion(0,1)!=0: 
            flag=0
            actualizacionPesos(1,0,0,1) 
        if funcionActivacion(1,0)!=0:
            flag=0
            actualizacionPesos(1,0,1,0) 
        if funcionActivacion(1,1)!=1: 
            flag=0
            actualizacionPesos(0,1,1,1)        
AND()
#graficar()

def OR():  
    flag=0
    while flag==0:
        flag = 1
        if funcionActivacion(0,0)!=0:
            flag=0
            actualizacionPesos(1,0,0,0)
        if funcionActivacion(0,1)!=1: 
            flag=0
            actualizacionPesos(0,1,0,1) 
        if funcionActivacion(1,0)!=1:
            flag=0
            actualizacionPesos(0,1,1,0) 
        if funcionActivacion(1,1)!=1: 
            flag=0
            actualizacionPesos(0,1,1,1)        
OR()
#graficar2()

def NOR():  
    flag=0
    while flag==0:
        flag = 1
        if funcionActivacion(0,0)!=1:
            flag=0
            actualizacionPesos(0,1,0,0)
        if funcionActivacion(0,1)!=0: 
            flag=0
            actualizacionPesos(1,0,0,1) 
        if funcionActivacion(1,0)!=0:
            flag=0
            actualizacionPesos(1,0,1,0) 
        if funcionActivacion(1,1)!=0: 
            flag=0
            actualizacionPesos(1,0,1,1)        
NOR()
#graficar3()
def NAND():  
    flag=0
    while flag==0:
        flag = 1
        if funcionActivacion(0,0)!=1:
            flag=0
            actualizacionPesos(0,1,0,0)
        if funcionActivacion(0,1)!=1: 
            flag=0
            actualizacionPesos(0,1,0,1) 
        if funcionActivacion(1,0)!=1:
            flag=0
            actualizacionPesos(0,1,1,0) 
        if funcionActivacion(1,1)!=0: 
            flag=0
            actualizacionPesos(1,0,1,1)        
NAND() 
#graficar4()
print ("Ingresa una compuerta AND, OR, NOR o NAND")
nombre=input()
if nombre=="AND":
    AND()
else:
    if nombre=="OR":
        OR()
    else:
        if nombre=="NAND":
            NAND()
        else:
            if nombre=="NOR":
                NOR()
                
def graficar():
    x1 = (b*(-1))/w1
    x2 = (b*(-1))/w2
    p1 = [x1,0]
    p2 = [0,x2]
    wx = [0,w1]
    wy = [0,w2]
    if nombre=="AND": 
        print("entro")
        plt.scatter(1,1,color='b')
        plt.scatter(0,1,color='r')
        plt.scatter(1,0,color='r')
        plt.scatter(0,0,color='r')
    else:
        if nombre=="OR":
            print("entro")
            plt.scatter(1,1,color='r')
            plt.scatter(0,1,color='r')
            plt.scatter(1,0,color='r')
            plt.scatter(0,0,color='b')
        else:
            if nombre=="NOR":
                plt.scatter(1,1,color='b')
                plt.scatter(0,1,color='b')
                plt.scatter(1,0,color='b')
                plt.scatter(0,0,color='r')
            else:
                if nombre=="NAND":
                    plt.scatter(1,1,color='B')
                    plt.scatter(0,1,color='r')
                    plt.scatter(1,0,color='r')
                    plt.scatter(0,0,color='R')
    plt.plot(p1,p2, color="black")
    plt.plot(wx,wy, color="red")
    plt.show()  
graficar()
```
